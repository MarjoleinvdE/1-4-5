```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
big5= pd.read_csv('big5.csv' , delimiter=';')
big5
```

```python


# List of personality traits
personality_traits = ['open', 'cons', 'extra', 'agree', 'neur']

# Create separate subplots for each personality trait
for trait in personality_traits:
    plt.figure(figsize=(6, 4))
    plt.hist(big5[trait], bins=10, color='skyblue', alpha=0.7)
    plt.xlabel(f'{trait.capitalize()} Scores')
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {trait.capitalize()} Scores')
    plt.grid(True)
    plt.show()



```

```python
eindtoests = pd.read_csv('eindtoets.csv',delimiter=';')
eindtoests
```

```python

plt.figure(figsize=(10, 6))

# Create a histogram for the 'standaardscore' column
plt.hist(eindtoests['standaardscore'], bins=50, color='skyblue', edgecolor='black', alpha=0.7)

# Add labels and title
plt.xlabel('Standard Scores')
plt.ylabel('Frequency')
plt.title('Distribution of Standard Scores')

# Show the plot
plt.grid(True)
plt.show()
```

In summary, the histogram illustrates a positive correlation between higher standard scores and increased frequency of students. As scores rise, more students achieve higher scores, indicating a positive trend in student performance.

```python


# Plot histograms for each subscale
plt.figure(figsize=(14, 5))

# Language Subscale (ond_opgaves8000) histogram
plt.subplot(1, 3, 1)
plt.hist(eindtoests['ond_opgaves8000'].dropna(), bins=15, color='skyblue', edgecolor='black')
plt.title('Language Subscale Questions Attempted')
plt.xlabel('Number of Questions')
plt.ylabel('Frequency')

# Match Subscale (ond_opgaves8001) histogram
plt.subplot(1, 3, 2)
plt.hist(eindtoests['ond_opgaves8001'].dropna(), bins=15, color='skyblue', edgecolor='black')
plt.title('Match Subscale Questions Attempted')
plt.xlabel('Number of Questions')
plt.ylabel('Frequency')

# Knowledge Subscale (ond_opgaves8002) histogram
plt.subplot(1, 3, 3)
plt.hist(eindtoests['ond_opgaves8002'].dropna(), bins=15, color='skyblue', edgecolor='black')
plt.title('Knowledge Subscale Questions Attempted')
plt.xlabel('Number of Questions')
plt.ylabel('Frequency')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
# Plot histograms for each subscale correct answer
plt.figure(figsize=(14, 5))

# Language Subscale correct answer (ond_percgoed8000) histogram
plt.subplot(1, 3, 1)
plt.hist(eindtoests['ond_percgoed8000'].dropna(), bins=15, color='skyblue', edgecolor='black')
plt.title('Correct answers in Language Subscale')
plt.xlabel('Number of correct aswer')
plt.ylabel('Frequency')

# Match Subscale correct answer(ond_percgoed8001) histogram
plt.subplot(1, 3, 2)
plt.hist(eindtoests['ond_percgoed8001'].dropna(), bins=15, color='skyblue', edgecolor='black')
plt.title('correct answers in Match  Subscale ')
plt.xlabel('Number of correct answer')
plt.ylabel('Frequency')

# Knowledge Subscale (ond_percgoed8002) histogram
plt.subplot(1, 3, 3)
plt.hist(eindtoests['ond_percgoed8002'].dropna(), bins=15, color='skyblue', edgecolor='black')
plt.title('correct answers Knowledge  Subscale')
plt.xlabel('Number of correct answer')
plt.ylabel('Frequency')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
