```python
#make the df now procesed_dataset.csv
df = pd.read_csv('processed_dataset.csv', delimiter=',')
df['timestampsm'] = df['timestampsm'].apply(lambda x: x/1000)


#only show patternsm, numbercl and timestampsm where timestamp is greater than variable x

x=180
df = df[df['timestampsm'] > x]

#only show patternsm, numbercl and timestampsm
df = df[['patternsm', 'numbercl', 'timestampsm', 'idll_vo3lv']]
df.head(200)

```

```python
df['statuscl_char_count'] = df['numbercl'].apply(lambda x: len(str(x).split(',')))
df.head(200)
#we count the number of characters in the numbercl column and put it in a new column called statuscl_char_count
```

```python
import pandas as pd

# Assuming df is your DataFrame

# Drop rows with NaN in 'numbercl'
df = df.dropna(subset=['numbercl'])

# Split 'numbercl' and convert to integers
df['numbercl'] = df['numbercl'].str.split(',').apply(lambda x: [int(i) for i in x])

# Check for duplicates in 'numbercl'
df['has_repeated_numbers'] = df['numbercl'].apply(lambda x: len(x) != len(set(x)))

# Filter rows with repeated numbers in 'numbercl'
filtered_df = df[df['has_repeated_numbers']]

filtered_df
```
