```python
import pandas as pd
import numpy as np

# 1. Read in the data
pd.set_option('display.max_rows', 200)
pd.set_option('display.max_columns', 200)
df = pd.read_csv('FivePoints.csv', delimiter=';')



pivot_df = pd.pivot_table(df, index=[ 'studentid', 'timestamp'], columns='name', values='value', aggfunc=lambda x: x).reset_index()
#make a new dataset, derived from pivot_df with only rows where there is a NaN value
pivot_df.head(100)
```



Combine sessions

make new data


```python
# drop the rows with NaN values in timestampload column and the row below it
pivot_df = pivot_df.dropna(subset=['timestampload'], axis=0).reset_index(drop=True)
pivot_df.head(100)
```

# MERGING ALL DATASETS


```python
bigfive_df = pd.read_csv('big5.csv', delimiter=';')
#sort the studentid ascending
bigfive_df = bigfive_df.sort_values(by=['studentid'])
bigfive_df
```

```python
merged_df = pd.merge(pivot_df, bigfive_df, on='studentid')
merged_df.head(100)
```

```python
#red the eindtoets.csv
eindtoets_df = pd.read_csv('eindtoets.csv', delimiter=';')
#sort it ascending by studentid
eindtoets_df = eindtoets_df.sort_values(by=['studentid'])
#merge the eindtoets with the merged_df where students are the same
merged_df = pd.merge(merged_df, eindtoets_df, on='studentid')
merged_df.head(100)
```

```python
#next we count the unique numbers of students
merged_df['studentid'].nunique()
#only show students that have a standaardscore of 545 or higher
#order the set by standaardscore descending
merged_df = merged_df.sort_values(by=['standaardscore'], ascending=False)
#count the NaN values
merged_df.isnull().sum()


#merged_df = merged_df[merged_df['standaardscore'] <= 511]
```
