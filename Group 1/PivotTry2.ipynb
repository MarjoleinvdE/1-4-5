```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 200)
df = pd.read_csv('FivePoints.csv', delimiter=';')
#next we count the number of unique studentid values
df['studentid'].nunique()
```

We have to change it to make it look like a "normal" dataset

```python

# Pivot the table
df_pivoted = df.pivot(index='primary', columns='name', values='value')


# Display the resulting DataFrame
df_pivoted.head(100)
```

Next, we merge them

```python
df_merged = pd.merge(df, df_pivoted, on='primary')
df_merged = df_merged.head(1000)
df_merged.head(100)
```

```python
df_new = df_merged

# Iterate through every 6 rows to process the data
for i in range(0, len(df_new) - 6, 6):
    subset = df_new.iloc[i:i+6]

    # Iterate through each column and move non-null values to the top
    for column in subset.columns:
        non_null_values = subset[column].dropna().tolist()
        subset[column] = non_null_values + [np.nan] * (6 - len(non_null_values))

    # Update the original DataFrame with the processed subset
    df_new.iloc[i:i+6] = subset

# Remove rows where all values are null
df_new.dropna(how='all', inplace=True)

# Reset the index
df_new.reset_index(drop=True, inplace=True)


df_new.head(100)
```

```python
#drop the columns that are not needed
df_new = df_new.drop(['name', 'value'], axis=1)
df_new.head(100)

#rename the columns

```
